```
#integration function
from scipy import integrate
help(integrate)


```

```
i=integrate.quad(lambda x:special.exp10(x),0,1)
print(i)
```

```
e=lambda x,y:x*y**2
f=lambda x:1
g=lambda x:-1
integrate.dblquad(e,0,1,f,g)
```

```
#Fourier transforms
from scipy import fft,ifft
import numpy as np
x=np.array([1,2,3,4])
y=fft(x)
print(y)
```

```
#Fourier transforms
from scipy import fft,ifft
import numpy as np
x=np.array([1,2,3,4])
y=ifft(x)
print(y)
```

```
#Linear algebra
from scipy import linalg
a=np.array([[1,2],[3,4]])
b=linalg.inv(a)
print(b)
```

```
#interpolation functions
import matplotlib.pyplot as plt
from scipy import interpolate
x=np.arange(5,20)
y=np.exp(x/3.0)
f=interpolate.interp1d(x,y)
x1=np.arange(6,12)
y1=f(x1)
plt.plot(x,y,'o',x1,y1,'--')
```
